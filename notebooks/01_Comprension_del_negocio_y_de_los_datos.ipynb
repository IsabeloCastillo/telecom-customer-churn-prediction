## Decisión 1 - Columna Cargos_Totales

Se detectaron 11 registros con valores vacíos. Tras investigar, se comprobó que todos correspondían a clientes con Antiguedad_Meses = 0. Se concluye que estos clientes aún no habían generado cargos acumulados. Por tanto, los valores vacíos se sustituirán por 0 antes de convertir la columna a tipo numérico.

```python
# Sustituir cadenas vacías por 0
df["Cargos_Totales"] = df["Cargos_Totales"].replace(" ", "0")

# Convertir la columna a tipo numérico
df["Cargos_Totales"] = pd.to_numeric(df["Cargos_Totales"])
```
